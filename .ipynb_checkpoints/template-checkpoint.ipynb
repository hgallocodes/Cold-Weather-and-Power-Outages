```python
# Combining times and setting datetime type

def combine_times(df, col1, col2):
    combine = df[col1].astype(str).str[:11] + df[col2].astype(str)
    return pd.to_datetime(combine.apply(lambda x: np.nan if x == 'nannan' else x), format='%Y-%m-%d %H:%M:%S')
outages = outages.assign(**{'OUTAGE.START':combine_times(outages, "OUTAGE.START.DATE", "OUTAGE.START.TIME")})
outages = outages.assign(**{'OUTAGE.RESTORATION':combine_times(outages, "OUTAGE.RESTORATION.DATE", "OUTAGE.RESTORATION.TIME")})
```

```python
def clean_categories(text):
    if text == ' Hydro' or text == 'Hydro':
        return 'hydro'
    elif text == ' Coal' or text == 'Coal':
        return 'coal'
    elif text == ' Natural Gas':
        return 'natural gas'
    elif text == 'Petroleum':
        return 'petroleum'
    else:
        return text
```

### Assessment of Missingness

```python
# TODO
def k_s_stat(df, col):
    first_group = df[df[col].notna()]['CAUSE.CATEGORY.DETAIL']
    second_group = df[df[col].isna()]['CAUSE.CATEGORY.DETAIL']
    return ks_2samp(first_group, second_group).statistic
```

```python
def get_pvals(df):

    columns = df.columns

    p_vals = []

    for i in columns:
        if i != 'CAUSE.CATEGORY.DETAIL':
            n = 10000
            
            ks = []
            
            for _ in range(n):
                df['shuffled_column'] = np.random.permutation(df[i])
                ks.append(k_s_stat(df,'shuffled_column'))

            p_val = (np.array(ks) >= k_s_stat(df, i)).mean()

            p_vals.append(p_val)

        else:
            continue
    return p_vals
```

```python
get_pvals(assess_mar)
```
